## Read Data

```python
data = pd.read_csv(DATA_PATH, header=0, index_col=0)
data
```

## Frequency Analysis

```python
col_y = "nodeWeight"
measure = "frequency"
visualization.show_measure(data, "Node", col_y, 
                           ascending=False, 
                           xlabel="objects",
                           ylabel=measure.capitalize(),
                           title=f"Object {measure} analysis")
visualization.show_distribution(data, col_y, 
                                xlabel=measure.capitalize(), 
                                y_label="density", 
                                title=f"The density of the {measure} of each node in the graph")

visualization.show_measure(data, "Node", col_y, 
                           grouped_by="type", 
                           ascending=False, 
                           xlabel="objects",
                           ylabel=measure.capitalize(),
                           title=f"Object {measure} analysis grouped by the type of the objects (thing/stuff)")

visualization.show_distribution(data, col_y, 
                                xlabel=measure.capitalize(), 
                                y_label="density", 
                                grouped_by="type",
                                title=f"The density of the {measure} of each node in the graph")
```
